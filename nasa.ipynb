```python
# ============================================================
# SECTION 1 - LOAD AND VALIDATE FD001
# Step 1: load the raw train, test, and RUL files
# ============================================================

train_path = DATA_DIR / "train_FD001.txt"
test_path = DATA_DIR / "test_FD001.txt"
rul_path = DATA_DIR / "RUL_FD001.txt"

train_rows = []
with train_path.open("r", encoding="utf-8") as file:
    for line in file:
        line = line.strip()
        if line:
            train_rows.append(tuple(float(value) for value in line.split()))

test_rows = []
with test_path.open("r", encoding="utf-8") as file:
    for line in file:
        line = line.strip()
        if line:
            test_rows.append(tuple(float(value) for value in line.split()))

rul_values = []
with rul_path.open("r", encoding="utf-8") as file:
    for line in file:
        line = line.strip()
        if line:
            rul_values.append(int(float(line)))

print("train rows loaded:", len(train_rows))
print("test rows loaded:", len(test_rows))
print("RUL values loaded:", len(rul_values))
```

```python
# ============================================================
# SECTION 1 - STRUCTURAL VALIDATION
# ============================================================

train_column_counts = sorted(set(len(row) for row in train_rows))
test_column_counts = sorted(set(len(row) for row in test_rows))

train_units = sorted({int(row[0]) for row in train_rows})
test_units = sorted({int(row[0]) for row in test_rows})

train_cycles_by_unit = {}
for row in train_rows:
    unit = int(row[0])
    cycle = int(row[1])
    if unit not in train_cycles_by_unit:
        train_cycles_by_unit[unit] = []
    train_cycles_by_unit[unit].append(cycle)

test_cycles_by_unit = {}
for row in test_rows:
    unit = int(row[0])
    cycle = int(row[1])
    if unit not in test_cycles_by_unit:
        test_cycles_by_unit[unit] = []
    test_cycles_by_unit[unit].append(cycle)

train_cycle_lengths = [max(cycles) for cycles in train_cycles_by_unit.values()]
test_cycle_lengths = [max(cycles) for cycles in test_cycles_by_unit.values()]

train_units_expected = list(range(1, len(train_units) + 1))
test_units_expected = list(range(1, len(test_units) + 1))

train_cycles_are_consecutive = True
for unit, cycles in train_cycles_by_unit.items():
    if cycles != list(range(1, len(cycles) + 1)):
        train_cycles_are_consecutive = False
        print("non-consecutive train cycles found for unit", unit)
        break

test_cycles_are_consecutive = True
for unit, cycles in test_cycles_by_unit.items():
    if cycles != list(range(1, len(cycles) + 1)):
        test_cycles_are_consecutive = False
        print("non-consecutive test cycles found for unit", unit)
        break

summary = {
    "columns": COLUMNS,
    "train_row_count": len(train_rows),
    "test_row_count": len(test_rows),
    "rul_count": len(rul_values),
    "train_column_counts": train_column_counts,
    "test_column_counts": test_column_counts,
    "train_engine_count": len(train_units),
    "test_engine_count": len(test_units),
    "train_cycle_lengths": {
        "min": min(train_cycle_lengths),
        "mean": round(mean(train_cycle_lengths), 2),
        "max": max(train_cycle_lengths),
    },
    "test_cycle_lengths": {
        "min": min(test_cycle_lengths),
        "mean": round(mean(test_cycle_lengths), 2),
        "max": max(test_cycle_lengths),
    },
    "rul_summary": {
        "min": min(rul_values),
        "mean": round(mean(rul_values), 2),
        "max": max(rul_values),
    },
    "train_units_are_consecutive": train_units == train_units_expected,
    "test_units_are_consecutive": test_units == test_units_expected,
    "train_cycles_are_consecutive": train_cycles_are_consecutive,
    "test_cycles_are_consecutive": test_cycles_are_consecutive,
}
```

```python
# ============================================================
# SECTION 3 - PREPROCESSING AND SENSOR SELECTION
# Choose which columns will actually be used by the model
# ============================================================

bookkeeping_cols = ["unit", "cycle"]

# For FD001, op3 is constant and op1/op2 vary only slightly around one regime.
# To keep the first version simple, we drop all three operational settings.
dropped_op_cols = ["op1", "op2", "op3"]

# From EDA: these sensors are constant or nearly constant in FD001.
dropped_sensor_cols = sorted(constant_sensors + nearly_constant_sensors)

model_feature_cols = [col for col in sensor_cols if col not in dropped_sensor_cols]
dropped_feature_cols = dropped_op_cols + dropped_sensor_cols

preprocessing_summary_df = pd.DataFrame(
    {
        "bookkeeping_columns": [", ".join(bookkeeping_cols)],
        "dropped_operational_settings": [", ".join(dropped_op_cols)],
        "dropped_sensors": [", ".join(dropped_sensor_cols)],
        "number_of_model_features": [len(model_feature_cols)],
        "kept_model_features": [", ".join(model_feature_cols)],
    }
)

print(preprocessing_summary_df.to_string(index=False))
```

```python
# Separate metadata from model features

train_meta_df = train_df[bookkeeping_cols].copy()
test_meta_df = test_df[bookkeeping_cols].copy()

train_features_raw_df = train_df[model_feature_cols].copy()
test_features_raw_df = test_df[model_feature_cols].copy()

print("train metadata shape:", train_meta_df.shape)
print("test metadata shape:", test_meta_df.shape)
print("train raw feature shape:", train_features_raw_df.shape)
print("test raw feature shape:", test_features_raw_df.shape)
```

```python
# Standardize using training statistics only
train_feature_means = train_features_raw_df.mean()
train_feature_stds = train_features_raw_df.std()
train_feature_stds[train_feature_stds == 0] = 1.0

train_features_scaled_df = (train_features_raw_df - train_feature_means) / train_feature_stds
test_features_scaled_df = (test_features_raw_df - train_feature_means) / train_feature_stds

scaling_summary_df = pd.DataFrame(
    {
        "train_mean_before_scaling": train_feature_means,
        "train_std_before_scaling": train_feature_stds,
        "train_mean_after_scaling": train_features_scaled_df.mean(),
        "train_std_after_scaling": train_features_scaled_df.std(),
    }
)

print(scaling_summary_df.round(4).head(10).to_string())
```

```python
# ============================================================
# SECTION 4 - RUL CONSTRUCTION FOR EVALUATION
# Build row-level RUL and a simple near-failure label
# ============================================================

failure_threshold_rul = 30

# TRAIN RUL: each training engine runs until failure.
train_max_cycles = train_df.groupby("unit")["cycle"].transform("max")
train_eval_df = train_meta_df.copy()
train_eval_df["RUL"] = train_max_cycles - train_df["cycle"]

# TEST RUL: RUL file gives remaining life after the last observed cycle.
test_last_cycles_df = test_df.groupby("unit", as_index=False)["cycle"].max().rename(columns={"cycle": "last_cycle"})
rul_df_with_unit = rul_df.copy()
rul_df_with_unit["unit"] = range(1, len(rul_df_with_unit) + 1)
rul_df_with_unit = rul_df_with_unit[["unit", "RUL"]].rename(columns={"RUL": "final_rul_after_last_cycle"})

test_rul_lookup_df = test_last_cycles_df.merge(rul_df_with_unit, on="unit", how="left")
test_eval_df = test_meta_df.merge(test_rul_lookup_df, on="unit", how="left")
test_eval_df["RUL"] = test_eval_df["final_rul_after_last_cycle"] + (test_eval_df["last_cycle"] - test_eval_df["cycle"])

train_eval_df["is_near_failure"] = (train_eval_df["RUL"] <= failure_threshold_rul).astype(int)
test_eval_df["is_near_failure"] = (test_eval_df["RUL"] <= failure_threshold_rul).astype(int)

print("failure threshold RUL:", failure_threshold_rul)
```

```python
# ============================================================
# SECTION 5 - CREATE SLIDING WINDOWS
# Step 5: convert row-by-row time series into short sequences
# ============================================================

import numpy as np

WINDOW_LENGTH = 30
STRIDE = 1
HEALTHY_MAX_CYCLE = 50

def create_sliding_windows(feature_df, eval_df, window_length=30, stride=1):
    """
    Create sliding windows per engine without mixing engines.

    Each window stores:
    - input values
    - engine ID
    - start cycle
    - end cycle
    - RUL at the end of the window
    - near-failure label at the end of the window
    """

    X_windows = []
    meta_rows = []

    combined_df = pd.concat(
        [
            eval_df[["unit", "cycle", "RUL", "is_near_failure"]].reset_index(drop=True),
            feature_df.reset_index(drop=True),
        ],
        axis=1,
    )

    feature_columns = feature_df.columns.tolist()

    for unit, unit_df in combined_df.groupby("unit", sort=True):
        unit_df = unit_df.sort_values("cycle").reset_index(drop=True)
        feature_values = unit_df[feature_columns].to_numpy(dtype=np.float32)

        for start_idx in range(0, len(unit_df) - window_length + 1, stride):
            end_idx = start_idx + window_length - 1

            window = feature_values[start_idx : start_idx + window_length]
            start_row = unit_df.iloc[start_idx]
            end_row = unit_df.iloc[end_idx]

            X_windows.append(window)
            meta_rows.append(
                {
                    "unit": int(unit),
                    "start_cycle": int(start_row["cycle"]),
                    "end_cycle": int(end_row["cycle"]),
                    "RUL": int(end_row["RUL"]),
                    "is_near_failure": int(end_row["is_near_failure"]),
                }
            )

    X = np.stack(X_windows).astype(np.float32)
    meta_df = pd.DataFrame(meta_rows)

    return X, meta_df


X_train_windows, train_window_meta_df = create_sliding_windows(
    train_features_scaled_df,
    train_eval_df,
    window_length=WINDOW_LENGTH,
    stride=STRIDE,
)

X_test_windows, test_window_meta_df = create_sliding_windows(
    test_features_scaled_df,
    test_eval_df,
    window_length=WINDOW_LENGTH,
    stride=STRIDE,
)

window_summary_df = pd.DataFrame(
    {
        "dataset": ["train_FD001", "test_FD001"],
        "windows": [len(X_train_windows), len(X_test_windows)],
        "window_shape": [str(X_train_windows.shape[1:]), str(X_test_windows.shape[1:])],
        "engines": [train_window_meta_df["unit"].nunique(), test_window_meta_df["unit"].nunique()],
        "window_length": [WINDOW_LENGTH, WINDOW_LENGTH],
        "stride": [STRIDE, STRIDE],
    }
)

print(window_summary_df.to_string(index=False))
print()
print("Train window metadata preview:")
print(train_window_meta_df.head().to_string(index=False))
print()
print("Test window metadata preview:")
print(test_window_meta_df.head().to_string(index=False))

assert X_train_windows.shape[1] == WINDOW_LENGTH
assert X_test_windows.shape[1] == WINDOW_LENGTH
assert X_train_windows.shape[2] == len(model_feature_cols)
assert X_test_windows.shape[2] == len(model_feature_cols)

assert train_window_meta_df[["unit", "start_cycle", "end_cycle", "RUL"]].isna().sum().sum() == 0
assert test_window_meta_df[["unit", "start_cycle", "end_cycle", "RUL"]].isna().sum().sum() == 0

assert (train_window_meta_df["end_cycle"] - train_window_meta_df["start_cycle"] + 1).eq(WINDOW_LENGTH).all()
assert (test_window_meta_df["end_cycle"] - test_window_meta_df["start_cycle"] + 1).eq(WINDOW_LENGTH).all()

print("\nStep 5 passed: sliding windows were created without mixing engines.")
```

```python
# ============================================================
# SECTION 6 - SELECT HEALTHY TRAINING WINDOWS
# Step 6: keep only early-life windows for anomaly-detector training
# ============================================================

# Healthy rule:
# A training window is considered healthy if its END cycle is within the first 50 cycles.
# With a 30-cycle window, this means windows ending at cycles 30, 31, ..., 50.

healthy_mask = train_window_meta_df["end_cycle"] <= HEALTHY_MAX_CYCLE

X_healthy_windows = X_train_windows[healthy_mask.to_numpy()]
healthy_window_meta_df = train_window_meta_df.loc[healthy_mask].reset_index(drop=True)

healthy_summary_df = pd.DataFrame(
    {
        "healthy_rule": [f"end_cycle <= {HEALTHY_MAX_CYCLE}"],
        "healthy_windows": [len(X_healthy_windows)],
        "engines_used": [healthy_window_meta_df["unit"].nunique()],
        "window_length": [WINDOW_LENGTH],
        "stride": [STRIDE],
        "features_per_cycle": [len(model_feature_cols)],
    }
)

print(healthy_summary_df.to_string(index=False))
print()
print("Healthy window metadata preview:")
print(healthy_window_meta_df.head().to_string(index=False))
print()
print(
    "Healthy end-cycle range:",
    int(healthy_window_meta_df["end_cycle"].min()),
    "to",
    int(healthy_window_meta_df["end_cycle"].max()),
)

assert len(X_healthy_windows) == 100 * (HEALTHY_MAX_CYCLE - WINDOW_LENGTH + 1)
assert healthy_window_meta_df["unit"].nunique() == 100
assert healthy_window_meta_df["end_cycle"].max() <= HEALTHY_MAX_CYCLE
assert healthy_window_meta_df["is_near_failure"].sum() == 0

print("\nStep 6 passed: healthy training windows are ready.")
```

```python
# ============================================================
# SECTION 7 - PCA RECONSTRUCTION ERROR BASELINE
# Step 7: simple anomaly detection baseline using PCA
# ============================================================

from sklearn.decomposition import PCA
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ------------------------------------------------------------
# 1. Flatten each window
# PCA expects 2D input:
# rows = windows
# columns = flattened sequence values
# Example: 30 cycles x 14 features = 420 columns
# ------------------------------------------------------------

X_healthy_flat = X_healthy_windows.reshape(X_healthy_windows.shape[0], -1)
X_train_flat = X_train_windows.reshape(X_train_windows.shape[0], -1)
X_test_flat = X_test_windows.reshape(X_test_windows.shape[0], -1)

print("Flattened healthy shape:", X_healthy_flat.shape)
print("Flattened train shape:  ", X_train_flat.shape)
print("Flattened test shape:   ", X_test_flat.shape)

# ------------------------------------------------------------
# 2. Fit PCA only on healthy windows
# n_components=0.95 means PCA keeps enough components to explain
# 95% of the variance in healthy behavior.
# ------------------------------------------------------------

pca_baseline = PCA(n_components=0.95, random_state=42)
pca_baseline.fit(X_healthy_flat)

print("\nPCA fitted successfully.")
print("Original flattened dimension:", X_healthy_flat.shape[1])
print("PCA components kept:", pca_baseline.n_components_)
print("Explained variance ratio:", round(pca_baseline.explained_variance_ratio_.sum(), 4))

# ------------------------------------------------------------
# 3. Reconstruction error function
# Higher reconstruction error = more abnormal behavior
# ------------------------------------------------------------

def pca_reconstruction_error(model, X_flat):
    X_reduced = model.transform(X_flat)
    X_reconstructed = model.inverse_transform(X_reduced)
    errors = np.mean((X_flat - X_reconstructed) ** 2, axis=1)
    return errors

healthy_pca_errors = pca_reconstruction_error(pca_baseline, X_healthy_flat)
train_pca_errors = pca_reconstruction_error(pca_baseline, X_train_flat)
test_pca_errors = pca_reconstruction_error(pca_baseline, X_test_flat)

# Add PCA anomaly scores to metadata
train_window_meta_df = train_window_meta_df.copy()
test_window_meta_df = test_window_meta_df.copy()
healthy_window_meta_df = healthy_window_meta_df.copy()

train_window_meta_df["pca_anomaly_score"] = train_pca_errors
test_window_meta_df["pca_anomaly_score"] = test_pca_errors
healthy_window_meta_df["pca_anomaly_score"] = healthy_pca_errors

# ------------------------------------------------------------
# 4. Basic PCA score summary
# ------------------------------------------------------------

pca_score_summary_df = pd.DataFrame(
    {
        "set": ["healthy_train", "all_train", "all_test"],
        "windows": [len(healthy_pca_errors), len(train_pca_errors), len(test_pca_errors)],
        "mean_error": [
            healthy_pca_errors.mean(),
            train_pca_errors.mean(),
            test_pca_errors.mean(),
        ],
        "median_error": [
            np.median(healthy_pca_errors),
            np.median(train_pca_errors),
            np.median(test_pca_errors),
        ],
        "max_error": [
            healthy_pca_errors.max(),
            train_pca_errors.max(),
            test_pca_errors.max(),
        ],
    }
)

print("\nPCA anomaly score summary:")
print(pca_score_summary_df.to_string(index=False))

# ------------------------------------------------------------
# 5. First simple benchmark using near-failure labels
# This is not the final evaluation yet, but it gives us a first idea.
# ------------------------------------------------------------

def safe_auc(y_true, scores, name):
    y_true = np.asarray(y_true)
    if len(np.unique(y_true)) < 2:
        print(f"{name} ROC-AUC: not available because only one class is present.")
        return np.nan

    auc = roc_auc_score(y_true, scores)
    print(f"{name} ROC-AUC: {auc:.4f}")
    return auc

print("\nFirst benchmark using is_near_failure labels:")
train_pca_auc = safe_auc(
    train_window_meta_df["is_near_failure"],
    train_window_meta_df["pca_anomaly_score"],
    "Train PCA",
)

test_pca_auc = safe_auc(
    test_window_meta_df["is_near_failure"],
    test_window_meta_df["pca_anomaly_score"],
    "Test PCA",
)

# ------------------------------------------------------------
# 6. Visualize PCA score distribution
# ------------------------------------------------------------

plt.figure(figsize=(10, 5))
plt.hist(healthy_pca_errors, bins=50, alpha=0.6, label="Healthy training windows")
plt.hist(train_pca_errors, bins=50, alpha=0.6, label="All training windows")
plt.xlabel("PCA reconstruction error")
plt.ylabel("Number of windows")
plt.title("PCA Reconstruction Error Distribution")
plt.legend()
plt.grid(True)
plt.show()

# ------------------------------------------------------------
# 7. Plot PCA anomaly score trend for a few training engines
# ------------------------------------------------------------

example_units = [1, 2, 3]

for unit in example_units:
    unit_scores = train_window_meta_df[train_window_meta_df["unit"] == unit]

    plt.figure(figsize=(10, 4))
    plt.plot(unit_scores["end_cycle"], unit_scores["pca_anomaly_score"])
    plt.xlabel("Cycle")
    plt.ylabel("PCA anomaly score")
    plt.title(f"PCA anomaly score over time - Train engine {unit}")
    plt.grid(True)
    plt.show()

print("\nStep 7 passed: PCA baseline anomaly scores were created.")
```

```python
# ============================================================
# SECTION 8 - LSTM AUTOENCODER
# Step 8: sequence-based anomaly detector
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, RepeatVector, TimeDistributed, Dense
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import roc_auc_score

# ------------------------------------------------------------
# 1. Reproducibility
# ------------------------------------------------------------

np.random.seed(42)
tf.random.set_seed(42)

print("TensorFlow version:", tf.__version__)
print("Healthy windows shape:", X_healthy_windows.shape)
print("Train windows shape:  ", X_train_windows.shape)
print("Test windows shape:   ", X_test_windows.shape)

# ------------------------------------------------------------
# 2. Define a small LSTM autoencoder
# Input shape = 30 cycles x 14 features
# ------------------------------------------------------------

timesteps = X_healthy_windows.shape[1]
n_features = X_healthy_windows.shape[2]

inputs = Input(shape=(timesteps, n_features))

# Encoder
encoded = LSTM(32, activation="tanh", return_sequences=False)(inputs)

# Bottleneck repeated over the sequence length
decoded = RepeatVector(timesteps)(encoded)

# Decoder
decoded = LSTM(32, activation="tanh", return_sequences=True)(decoded)

# Reconstruct the original sequence
outputs = TimeDistributed(Dense(n_features))(decoded)

lstm_autoencoder = Model(inputs, outputs)

lstm_autoencoder.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss="mse"
)

lstm_autoencoder.summary()

# ------------------------------------------------------------
# 3. Train only on healthy windows
# ------------------------------------------------------------

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True
)

history = lstm_autoencoder.fit(
    X_healthy_windows,
    X_healthy_windows,
    epochs=40,
    batch_size=64,
    validation_split=0.2,
    callbacks=[early_stop],
    shuffle=True,
    verbose=1
)

# ------------------------------------------------------------
# 4. Plot training history
# ------------------------------------------------------------

plt.figure(figsize=(8, 4))
plt.plot(history.history["loss"], label="Training loss")
plt.plot(history.history["val_loss"], label="Validation loss")
plt.xlabel("Epoch")
plt.ylabel("MSE loss")
plt.title("LSTM Autoencoder Training Loss")
plt.legend()
plt.grid(True)
plt.show()

# ------------------------------------------------------------
# 5. Compute LSTM reconstruction errors
# ------------------------------------------------------------

def lstm_reconstruction_error(model, X):
    X_reconstructed = model.predict(X, batch_size=256, verbose=0)
    errors = np.mean((X - X_reconstructed) ** 2, axis=(1, 2))
    return errors

healthy_lstm_errors = lstm_reconstruction_error(lstm_autoencoder, X_healthy_windows)
train_lstm_errors = lstm_reconstruction_error(lstm_autoencoder, X_train_windows)
test_lstm_errors = lstm_reconstruction_error(lstm_autoencoder, X_test_windows)

# Add LSTM scores to metadata
healthy_window_meta_df = healthy_window_meta_df.copy()
train_window_meta_df = train_window_meta_df.copy()
test_window_meta_df = test_window_meta_df.copy()

healthy_window_meta_df["lstm_anomaly_score"] = healthy_lstm_errors
train_window_meta_df["lstm_anomaly_score"] = train_lstm_errors
test_window_meta_df["lstm_anomaly_score"] = test_lstm_errors

# ------------------------------------------------------------
# 6. LSTM score summary
# ------------------------------------------------------------

lstm_score_summary_df = pd.DataFrame(
    {
        "set": ["healthy_train", "all_train", "all_test"],
        "windows": [len(healthy_lstm_errors), len(train_lstm_errors), len(test_lstm_errors)],
        "mean_error": [
            healthy_lstm_errors.mean(),
            train_lstm_errors.mean(),
            test_lstm_errors.mean(),
        ],
        "median_error": [
            np.median(healthy_lstm_errors),
            np.median(train_lstm_errors),
            np.median(test_lstm_errors),
        ],
        "max_error": [
            healthy_lstm_errors.max(),
            train_lstm_errors.max(),
            test_lstm_errors.max(),
        ],
    }
)

print("\nLSTM anomaly score summary:")
print(lstm_score_summary_df.to_string(index=False))

# ------------------------------------------------------------
# 7. First benchmark using near-failure labels
# ------------------------------------------------------------

def safe_auc(y_true, scores, name):
    y_true = np.asarray(y_true)

    if len(np.unique(y_true)) < 2:
        print(f"{name} ROC-AUC: not available because only one class is present.")
        return np.nan

    auc = roc_auc_score(y_true, scores)
    print(f"{name} ROC-AUC: {auc:.4f}")
    return auc

print("\nFirst benchmark using is_near_failure labels:")

train_lstm_auc = safe_auc(
    train_window_meta_df["is_near_failure"],
    train_window_meta_df["lstm_anomaly_score"],
    "Train LSTM",
)

test_lstm_auc = safe_auc(
    test_window_meta_df["is_near_failure"],
    test_window_meta_df["lstm_anomaly_score"],
    "Test LSTM",
)

# ------------------------------------------------------------
# 8. Compare PCA and LSTM quickly
# ------------------------------------------------------------

comparison_df = pd.DataFrame(
    {
        "model": ["PCA baseline", "LSTM autoencoder"],
        "train_auc": [train_pca_auc, train_lstm_auc],
        "test_auc": [test_pca_auc, test_lstm_auc],
    }
)

print("\nModel comparison:")
print(comparison_df.to_string(index=False))

# ------------------------------------------------------------
# 9. Plot LSTM anomaly score distribution
# ------------------------------------------------------------

plt.figure(figsize=(10, 5))
plt.hist(healthy_lstm_errors, bins=50, alpha=0.6, label="Healthy training windows")
plt.hist(train_lstm_errors, bins=50, alpha=0.6, label="All training windows")
plt.xlabel("LSTM reconstruction error")
plt.ylabel("Number of windows")
plt.title("LSTM Reconstruction Error Distribution")
plt.legend()
plt.grid(True)
plt.show()

# ------------------------------------------------------------
# 10. Plot LSTM score trend for example engines
# ------------------------------------------------------------

example_units = [1, 2, 3]

for unit in example_units:
    unit_scores = train_window_meta_df[train_window_meta_df["unit"] == unit]

    plt.figure(figsize=(10, 4))
    plt.plot(unit_scores["end_cycle"], unit_scores["lstm_anomaly_score"])
    plt.xlabel("Cycle")
    plt.ylabel("LSTM anomaly score")
    plt.title(f"LSTM anomaly score over time - Train engine {unit}")
    plt.grid(True)
    plt.show()

print("\nStep 8 passed: LSTM autoencoder anomaly scores were created.")
```

```python
# ============================================================
# SECTION 9 - CHOOSE AN ANOMALY THRESHOLD
# Step 9: convert anomaly scores into detection rules
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# 1. Compute threshold candidates from healthy-window errors
# ------------------------------------------------------------

def threshold_candidates(errors, model_name):
    """
    Compute simple anomaly thresholds from healthy-window reconstruction errors.
    """
    errors = np.asarray(errors)

    return pd.DataFrame(
        {
            "model": [model_name, model_name, model_name],
            "method": ["95th percentile", "99th percentile", "mean + 3 std"],
            "threshold": [
                np.percentile(errors, 95),
                np.percentile(errors, 99),
                errors.mean() + 3 * errors.std(),
            ],
            "healthy_false_alarm_rate": [
                np.mean(errors > np.percentile(errors, 95)),
                np.mean(errors > np.percentile(errors, 99)),
                np.mean(errors > (errors.mean() + 3 * errors.std())),
            ],
        }
    )


pca_threshold_df = threshold_candidates(
    healthy_window_meta_df["pca_anomaly_score"],
    "PCA baseline"
)

lstm_threshold_df = threshold_candidates(
    healthy_window_meta_df["lstm_anomaly_score"],
    "LSTM autoencoder"
)

threshold_table_df = pd.concat(
    [pca_threshold_df, lstm_threshold_df],
    ignore_index=True
)

print("Threshold candidates based only on healthy training windows:")
print(threshold_table_df.to_string(index=False))

# ------------------------------------------------------------
# 2. Select final threshold
# ------------------------------------------------------------
# We choose the 99th percentile because it is stricter than 95%.
# This means only around 1% of healthy windows are expected to be falsely flagged.
# This is a good choice for predictive maintenance because we want fewer false alarms.

pca_threshold = pca_threshold_df.loc[
    pca_threshold_df["method"] == "99th percentile", "threshold"
].iloc[0]

lstm_threshold = lstm_threshold_df.loc[
    lstm_threshold_df["method"] == "99th percentile", "threshold"
].iloc[0]

print("\nSelected final thresholds:")
print(f"PCA threshold:  {pca_threshold:.6f}")
print(f"LSTM threshold: {lstm_threshold:.6f}")

# ------------------------------------------------------------
# 3. Add raw anomaly flags
# ------------------------------------------------------------
# Detection rule:
# if anomaly_score > threshold, then the window is flagged as anomalous.

train_window_meta_df = train_window_meta_df.copy()
test_window_meta_df = test_window_meta_df.copy()
healthy_window_meta_df = healthy_window_meta_df.copy()

train_window_meta_df["pca_anomaly_flag"] = (
    train_window_meta_df["pca_anomaly_score"] > pca_threshold
).astype(int)

test_window_meta_df["pca_anomaly_flag"] = (
    test_window_meta_df["pca_anomaly_score"] > pca_threshold
).astype(int)

healthy_window_meta_df["pca_anomaly_flag"] = (
    healthy_window_meta_df["pca_anomaly_score"] > pca_threshold
).astype(int)

train_window_meta_df["lstm_anomaly_flag"] = (
    train_window_meta_df["lstm_anomaly_score"] > lstm_threshold
).astype(int)

test_window_meta_df["lstm_anomaly_flag"] = (
    test_window_meta_df["lstm_anomaly_score"] > lstm_threshold
).astype(int)

healthy_window_meta_df["lstm_anomaly_flag"] = (
    healthy_window_meta_df["lstm_anomaly_score"] > lstm_threshold
).astype(int)

# ------------------------------------------------------------
# 4. Optional false-alarm reduction: require consecutive alarms
# ------------------------------------------------------------
# A single high score may be noise.
# So we also create a stricter operational rule:
# anomaly is confirmed only after 3 consecutive windows exceed the threshold.

CONSECUTIVE_WINDOWS_REQUIRED = 3

def add_consecutive_alarm_flag(df, score_col, threshold, output_col, consecutive_windows=3):
    """
    Adds a flag that becomes 1 only when a unit has consecutive windows
    above the threshold.
    """
    df = df.copy()
    df[output_col] = 0

    for unit, unit_df in df.groupby("unit", sort=True):
        unit_df = unit_df.sort_values("end_cycle")
        above_threshold = (unit_df[score_col].to_numpy() > threshold)

        confirmed_flags = []
        counter = 0

        for is_above in above_threshold:
            if is_above:
                counter += 1
            else:
                counter = 0

            confirmed_flags.append(int(counter >= consecutive_windows))

        df.loc[unit_df.index, output_col] = confirmed_flags

    return df


train_window_meta_df = add_consecutive_alarm_flag(
    train_window_meta_df,
    score_col="pca_anomaly_score",
    threshold=pca_threshold,
    output_col="pca_confirmed_anomaly_flag",
    consecutive_windows=CONSECUTIVE_WINDOWS_REQUIRED,
)

test_window_meta_df = add_consecutive_alarm_flag(
    test_window_meta_df,
    score_col="pca_anomaly_score",
    threshold=pca_threshold,
    output_col="pca_confirmed_anomaly_flag",
    consecutive_windows=CONSECUTIVE_WINDOWS_REQUIRED,
)

train_window_meta_df = add_consecutive_alarm_flag(
    train_window_meta_df,
    score_col="lstm_anomaly_score",
    threshold=lstm_threshold,
    output_col="lstm_confirmed_anomaly_flag",
    consecutive_windows=CONSECUTIVE_WINDOWS_REQUIRED,
)

test_window_meta_df = add_consecutive_alarm_flag(
    test_window_meta_df,
    score_col="lstm_anomaly_score",
    threshold=lstm_threshold,
    output_col="lstm_confirmed_anomaly_flag",
    consecutive_windows=CONSECUTIVE_WINDOWS_REQUIRED,
)

# ------------------------------------------------------------
# 5. Detection summary
# ------------------------------------------------------------

def detection_summary(df, dataset_name):
    rows = []

    for model_name, raw_col, confirmed_col in [
        ("PCA baseline", "pca_anomaly_flag", "pca_confirmed_anomaly_flag"),
        ("LSTM autoencoder", "lstm_anomaly_flag", "lstm_confirmed_anomaly_flag"),
    ]:
        rows.append(
            {
                "dataset": dataset_name,
                "model": model_name,
                "windows": len(df),
                "raw_anomaly_windows": int(df[raw_col].sum()),
                "raw_anomaly_rate": df[raw_col].mean(),
                "confirmed_anomaly_windows": int(df[confirmed_col].sum()),
                "confirmed_anomaly_rate": df[confirmed_col].mean(),
            }
        )

    return pd.DataFrame(rows)


threshold_detection_summary_df = pd.concat(
    [
        detection_summary(train_window_meta_df, "train_FD001"),
        detection_summary(test_window_meta_df, "test_FD001"),
    ],
    ignore_index=True,
)

print("\nDetection summary using selected thresholds:")
print(threshold_detection_summary_df.to_string(index=False))

# ------------------------------------------------------------
# 6. Plot PCA and LSTM threshold behavior for example engines
# ------------------------------------------------------------

example_units = [1, 2, 3]

for unit in example_units:
    unit_scores = train_window_meta_df[train_window_meta_df["unit"] == unit]

    plt.figure(figsize=(10, 4))
    plt.plot(unit_scores["end_cycle"], unit_scores["pca_anomaly_score"], label="PCA anomaly score")
    plt.axhline(pca_threshold, linestyle="--", label="PCA threshold")
    plt.xlabel("Cycle")
    plt.ylabel("PCA anomaly score")
    plt.title(f"PCA threshold detection - Train engine {unit}")
    plt.legend()
    plt.grid(True)
    plt.show()

for unit in example_units:
    unit_scores = train_window_meta_df[train_window_meta_df["unit"] == unit]

    plt.figure(figsize=(10, 4))
    plt.plot(unit_scores["end_cycle"], unit_scores["lstm_anomaly_score"], label="LSTM anomaly score")
    plt.axhline(lstm_threshold, linestyle="--", label="LSTM threshold")
    plt.xlabel("Cycle")
    plt.ylabel("LSTM anomaly score")
    plt.title(f"LSTM threshold detection - Train engine {unit}")
    plt.legend()
    plt.grid(True)
    plt.show()

# ------------------------------------------------------------
# 7. Final detection rule statement
# ------------------------------------------------------------

print("\nFinal detection rule:")
print(f"PCA:  anomaly if PCA reconstruction error > {pca_threshold:.6f}")
print(f"LSTM: anomaly if LSTM reconstruction error > {lstm_threshold:.6f}")
print(
    f"Operational rule: confirm anomaly only after "
    f"{CONSECUTIVE_WINDOWS_REQUIRED} consecutive windows above the threshold."
)

print("\nStep 9 passed: anomaly thresholds and detection flags were created.")
```

```python
# ============================================================
# SECTION 10 - EVALUATE BASELINE AND LSTM
# Step 10: compare PCA and LSTM using metrics, plots, and lead time
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import (
    roc_auc_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
)

# ------------------------------------------------------------
# 1. Helper functions
# ------------------------------------------------------------

def evaluate_model_detection(df, dataset_name, model_name, score_col, raw_flag_col, confirmed_flag_col):
    """
    Evaluate anomaly scores and anomaly flags against near-failure labels.
    """

    y_true = df["is_near_failure"].astype(int).to_numpy()
    scores = df[score_col].to_numpy()
    raw_flags = df[raw_flag_col].astype(int).to_numpy()
    confirmed_flags = df[confirmed_flag_col].astype(int).to_numpy()

    # ROC-AUC uses continuous anomaly scores
    auc = roc_auc_score(y_true, scores) if len(np.unique(y_true)) > 1 else np.nan

    # Classification metrics use confirmed anomaly flags
    precision = precision_score(y_true, confirmed_flags, zero_division=0)
    recall = recall_score(y_true, confirmed_flags, zero_division=0)
    f1 = f1_score(y_true, confirmed_flags, zero_division=0)

    tn, fp, fn, tp = confusion_matrix(y_true, confirmed_flags, labels=[0, 1]).ravel()

    false_alarm_rate = fp / (fp + tn) if (fp + tn) > 0 else np.nan

    return {
        "dataset": dataset_name,
        "model": model_name,
        "windows": len(df),
        "ROC_AUC": auc,
        "precision": precision,
        "recall": recall,
        "F1_score": f1,
        "false_alarm_rate": false_alarm_rate,
        "TP": tp,
        "FP": fp,
        "TN": tn,
        "FN": fn,
        "raw_alarm_rate": raw_flags.mean(),
        "confirmed_alarm_rate": confirmed_flags.mean(),
    }


# ------------------------------------------------------------
# 2. Evaluation table for PCA and LSTM
# ------------------------------------------------------------

evaluation_rows = []

for dataset_name, df in [
    ("train_FD001", train_window_meta_df),
    ("test_FD001", test_window_meta_df),
]:
    evaluation_rows.append(
        evaluate_model_detection(
            df=df,
            dataset_name=dataset_name,
            model_name="PCA baseline",
            score_col="pca_anomaly_score",
            raw_flag_col="pca_anomaly_flag",
            confirmed_flag_col="pca_confirmed_anomaly_flag",
        )
    )

    evaluation_rows.append(
        evaluate_model_detection(
            df=df,
            dataset_name=dataset_name,
            model_name="LSTM autoencoder",
            score_col="lstm_anomaly_score",
            raw_flag_col="lstm_anomaly_flag",
            confirmed_flag_col="lstm_confirmed_anomaly_flag",
        )
    )

evaluation_df = pd.DataFrame(evaluation_rows)

print("Model evaluation summary:")
print(evaluation_df.to_string(index=False))


# ------------------------------------------------------------
# 3. Lead time calculation
# ------------------------------------------------------------
# Lead time = RUL at the first confirmed anomaly.
# Example:
# If the first confirmed anomaly appears when RUL = 40,
# the model gives around 40 cycles of warning.

def calculate_lead_time(df, dataset_name, model_name, confirmed_flag_col):
    """
    Calculate first detection cycle and lead time for each engine.
    """

    rows = []

    for unit, unit_df in df.groupby("unit", sort=True):
        unit_df = unit_df.sort_values("end_cycle").reset_index(drop=True)
        detected_df = unit_df[unit_df[confirmed_flag_col] == 1]

        if len(detected_df) == 0:
            rows.append(
                {
                    "dataset": dataset_name,
                    "model": model_name,
                    "unit": int(unit),
                    "detected": 0,
                    "first_detection_cycle": np.nan,
                    "lead_time_cycles": np.nan,
                    "rul_at_detection": np.nan,
                }
            )
        else:
            first_detection = detected_df.iloc[0]

            rows.append(
                {
                    "dataset": dataset_name,
                    "model": model_name,
                    "unit": int(unit),
                    "detected": 1,
                    "first_detection_cycle": int(first_detection["end_cycle"]),
                    "lead_time_cycles": int(first_detection["RUL"]),
                    "rul_at_detection": int(first_detection["RUL"]),
                }
            )

    return pd.DataFrame(rows)


lead_time_df = pd.concat(
    [
        calculate_lead_time(
            train_window_meta_df,
            "train_FD001",
            "PCA baseline",
            "pca_confirmed_anomaly_flag",
        ),
        calculate_lead_time(
            train_window_meta_df,
            "train_FD001",
            "LSTM autoencoder",
            "lstm_confirmed_anomaly_flag",
        ),
        calculate_lead_time(
            test_window_meta_df,
            "test_FD001",
            "PCA baseline",
            "pca_confirmed_anomaly_flag",
        ),
        calculate_lead_time(
            test_window_meta_df,
            "test_FD001",
            "LSTM autoencoder",
            "lstm_confirmed_anomaly_flag",
        ),
    ],
    ignore_index=True,
)

lead_time_summary_df = (
    lead_time_df
    .groupby(["dataset", "model"])
    .agg(
        engines=("unit", "count"),
        detected_engines=("detected", "sum"),
        detection_rate=("detected", "mean"),
        mean_lead_time=("lead_time_cycles", "mean"),
        median_lead_time=("lead_time_cycles", "median"),
        min_lead_time=("lead_time_cycles", "min"),
        max_lead_time=("lead_time_cycles", "max"),
    )
    .reset_index()
)

print("\nLead time summary:")
print(lead_time_summary_df.to_string(index=False))

print("\nLead time preview:")
print(lead_time_df.head(10).to_string(index=False))


# ------------------------------------------------------------
# 4. Compare PCA and LSTM side by side
# ------------------------------------------------------------

comparison_simple_df = evaluation_df[
    [
        "dataset",
        "model",
        "ROC_AUC",
        "precision",
        "recall",
        "F1_score",
        "false_alarm_rate",
        "confirmed_alarm_rate",
    ]
].copy()

print("\nSimple model comparison:")
print(comparison_simple_df.to_string(index=False))


# ------------------------------------------------------------
# 5. Plot anomaly scores over time with threshold lines
# ------------------------------------------------------------

def plot_model_scores_for_units(df, dataset_name, units, model_name, score_col, threshold, confirmed_flag_col):
    """
    Plot anomaly score over time for selected engines.
    """

    for unit in units:
        unit_scores = df[df["unit"] == unit].sort_values("end_cycle")

        plt.figure(figsize=(10, 4))
        plt.plot(unit_scores["end_cycle"], unit_scores[score_col], label=f"{model_name} anomaly score")
        plt.axhline(threshold, linestyle="--", label="Threshold")

        # Mark where near-failure starts, if it exists for this unit
        near_failure_rows = unit_scores[unit_scores["is_near_failure"] == 1]
        if len(near_failure_rows) > 0:
            near_failure_start = near_failure_rows["end_cycle"].min()
            plt.axvline(near_failure_start, linestyle=":", label="Near-failure starts")

        # Mark first confirmed detection
        detected_rows = unit_scores[unit_scores[confirmed_flag_col] == 1]
        if len(detected_rows) > 0:
            first_detection_cycle = detected_rows["end_cycle"].iloc[0]
            plt.axvline(first_detection_cycle, linestyle="-.", label="First confirmed detection")

        plt.xlabel("Cycle")
        plt.ylabel("Anomaly score")
        plt.title(f"{model_name} anomaly detection - {dataset_name} engine {unit}")
        plt.legend()
        plt.grid(True)
        plt.show()


example_units = [1, 2, 3]

plot_model_scores_for_units(
    train_window_meta_df,
    dataset_name="train_FD001",
    units=example_units,
    model_name="PCA baseline",
    score_col="pca_anomaly_score",
    threshold=pca_threshold,
    confirmed_flag_col="pca_confirmed_anomaly_flag",
)

plot_model_scores_for_units(
    train_window_meta_df,
    dataset_name="train_FD001",
    units=example_units,
    model_name="LSTM autoencoder",
    score_col="lstm_anomaly_score",
    threshold=lstm_threshold,
    confirmed_flag_col="lstm_confirmed_anomaly_flag",
)


# ------------------------------------------------------------
# 6. Bar chart comparison of ROC-AUC
# ------------------------------------------------------------

for dataset_name in evaluation_df["dataset"].unique():
    subset = evaluation_df[evaluation_df["dataset"] == dataset_name]

    plt.figure(figsize=(7, 4))
    plt.bar(subset["model"], subset["ROC_AUC"])
    plt.ylim(0, 1)
    plt.ylabel("ROC-AUC")
    plt.title(f"ROC-AUC comparison - {dataset_name}")
    plt.grid(axis="y")
    plt.show()


# ------------------------------------------------------------
# 7. Bar chart comparison of false alarm rate
# ------------------------------------------------------------

for dataset_name in evaluation_df["dataset"].unique():
    subset = evaluation_df[evaluation_df["dataset"] == dataset_name]

    plt.figure(figsize=(7, 4))
    plt.bar(subset["model"], subset["false_alarm_rate"])
    plt.ylim(0, 1)
    plt.ylabel("False alarm rate")
    plt.title(f"False alarm comparison - {dataset_name}")
    plt.grid(axis="y")
    plt.show()


# ------------------------------------------------------------
# 8. Lead time distribution
# ------------------------------------------------------------

detected_lead_times = lead_time_df[lead_time_df["detected"] == 1]

for dataset_name in detected_lead_times["dataset"].unique():
    subset = detected_lead_times[detected_lead_times["dataset"] == dataset_name]

    plt.figure(figsize=(8, 4))

    for model_name in subset["model"].unique():
        model_subset = subset[subset["model"] == model_name]
        plt.hist(
            model_subset["lead_time_cycles"].dropna(),
            bins=20,
            alpha=0.6,
            label=model_name,
        )

    plt.xlabel("Lead time in cycles")
    plt.ylabel("Number of engines")
    plt.title(f"Lead time distribution - {dataset_name}")
    plt.legend()
    plt.grid(True)
    plt.show()


# ------------------------------------------------------------
# 9. Short automatic interpretation
# ------------------------------------------------------------

best_train_model = evaluation_df[evaluation_df["dataset"] == "train_FD001"].sort_values(
    "ROC_AUC", ascending=False
).iloc[0]

best_test_model = evaluation_df[evaluation_df["dataset"] == "test_FD001"].sort_values(
    "ROC_AUC", ascending=False
).iloc[0]

print("\nAutomatic interpretation:")
print(
    f"On the training windows, the best ROC-AUC is achieved by "
    f"{best_train_model['model']} with ROC-AUC = {best_train_model['ROC_AUC']:.4f}."
)

print(
    f"On the test windows, the best ROC-AUC is achieved by "
    f"{best_test_model['model']} with ROC-AUC = {best_test_model['ROC_AUC']:.4f}."
)

print(
    "Lead time measures how many cycles before failure the first confirmed anomaly is detected."
)

print("\nStep 10 passed: PCA and LSTM were evaluated and compared.")
```
